```python
class sensor:
    def __init__(self, random, agentlocation):
        self.random = random
        self.agentlocation = agentlocation
        
    def sensorRoll(percept, agentlocation):
        roll = self.random.RandomRoll(9)
        if roll == 8:
            percepts = percept.getPercepts(agentlocation)
            percepts = setEmptyShelf(percepts)
            return percept
            
        elif roll == 9:
            percepts = percept.getPercepts(agentlocation)
            percepts = setShelf(percept)
            return percepts
        else:
            percepts = percept.getPercepts(agentlocation)
            return percepts
    
    #need the percept to be an array of dictionaries
    def setEmptyShelf(percept):
        shelves = []
        for (i in percept):
            if (percept[i] == 3):
                shelves.append(i)
            
        roll = self.random.RandomRoll(len(shelves))
        key = shelves[roll]
        percept[key] = -1
        return percept    
    
    def setShelf(percept):
        empty = []
        for (i in percept):
            if (percept[i] == -1):
                empty.append(i)
            
        roll = self.random.RandomRoll(len(empty))
        key = empty[roll]
        percept[key] = 3
        return percept 
```
